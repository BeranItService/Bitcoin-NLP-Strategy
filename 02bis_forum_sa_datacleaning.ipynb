# Forum data sentiment analysis results cleaning and analysing
In this notebook we carry out data cleaning and useful groupbys on a private data source containing comments from a forum about bitcoin.

```python
import pandas as pd
import numpy as np
#import nltk
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
pd.set_option('display.max_rows', 15)
```

```python
data= pd.read_csv('Data/Bigger Data/forum_NLP.csv')
data = data.drop('Unnamed: 0', axis = 1)
```

```python
data['merit_compound'] = data["merit"]*pd.to_numeric(data["compound"])
```

```python
data["newdate"]=data["datetime"]
```

```python
data = data[data["newdate"] < "2018-05-01"]
```

```python
data['newdate'] =  pd.to_datetime(data['newdate'], format='%Y%m%d %H:%M:%S')
```

```python
data['newdate'] = data['newdate'].dt.date
```

```python
data
```

```python
#df = data.iloc[:,[6,7,8,9,10,11]]
df = data.iloc[:,[6,10,11]]
df = df.groupby('newdate', as_index=False, sort=False).mean()
print (df)
```
