# Backtesting an investment strategy based on model predictions
In this notebook we will compare the performance of a buy/sell strategy according to the predictions of opur model versus a simple buy and hold strategy.

```python
import pandas as pd
import numpy as np
import ffn

%matplotlib inline


pd.set_option('display.max_rows', 15)
```

```python
prices = prices[["close","date"]]
```

## Predictions

```python
predictions = pd.read_csv("Data/nonlagged_predictions.csv", index_col = 0)
```

```python
predictions
```

## Merge tables

```python
data = pd.concat((prices, predictions), axis = 1)
data.to_csv("Data/backtest_data.csv")
data
```

## Backtesting the investment strategy

```python
btc_price = data[["close","predict"]]
btc_price[["close"]] = btc_price[["close"]].rebase(100)
btc_price
```

### Long & Out strategy:

```python
btc_returns = btc_price.copy()
btc_returns["returns"] = btc_returns["close"].pct_change()
btc_returns["returns"] = btc_returns["returns"] #.shift(-1)
#btc_returns["strategy"] = btc_returns["close"]
#btc_returns["strategy"].loc[data[["predict"]] == False, btc_returns["strategy"]] = 0
btc_returns["strategy_returns"] = btc_returns["returns"]
btc_returns.loc[btc_returns["predict"] == False, "strategy_returns"] = 0
```

```python
btc_returns["strategy_close"] = btc_returns["close"]
btc_returns["strategy_close"] = btc_returns["close"].iloc[0] * np.cumprod(btc_returns["strategy_returns"] + 1)
```

```python
btc_returns.index = pd.to_datetime(btc_returns.index)
```

## Strategy 1: Long & Out versus Long only
*based on predictions recommendation versus long position*

```python
performance_comparison = btc_returns[["close","strategy_close"]]
performance_comparison.columns = ["Passive Strategy","Active Long & Out Strategy"]
performance_comparison.plot(figsize = (15,9));
```

```python
stats = performance_comparison.calc_stats()
stats.display()
```

### Long & short strategy:

```python
btc_returns = btc_price.copy()
btc_returns["returns"] = btc_returns["close"].pct_change()
btc_returns["returns"] = btc_returns["returns"] #.shift(-1)
#btc_returns["strategy"] = btc_returns["close"]
#btc_returns["strategy"].loc[data[["predict"]] == False, btc_returns["strategy"]] = 0
btc_returns["strategy_returns"] = btc_returns["returns"]
btc_returns.loc[btc_returns["predict"] == False, "strategy_returns"] = -btc_returns["strategy_returns"]
```

```python
btc_returns["strategy_close"] = btc_returns["close"]
btc_returns["strategy_close"] = btc_returns["close"].iloc[0] * np.cumprod(btc_returns["strategy_returns"] + 1)
```

```python
btc_returns.index = pd.to_datetime(btc_returns.index)
```

# Strategy 2: Long & Short versus Long only
*based on predictions recommendations*

```python
performance_comparison = btc_returns[["close","strategy_close"]]
performance_comparison.columns = ["Passive Strategy","Active Long & Short Strategy"]
performance_comparison.plot(figsize = (15,9));
```

```python
stats = performance_comparison.calc_stats()
stats.display()
```
